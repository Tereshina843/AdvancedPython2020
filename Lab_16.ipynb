```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Задача 1

Our function:

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)
```

## Приближение многочленом первой степени

```python
x_1 = np.array([1,15])
A_1 = np.array([[1,1],[1,15]])
b_1 = np.zeros(2)
b_1 = f(x_1)
w_1 = np.zeros(2) #polynomial coefficients 
w_1 = np.linalg.solve(A_1, b_1)
reversed_w_1 = w_1[::-1]
p_1 = np.poly1d(reversed_w_1) #polynom
print(p_1)
```

```python
x_1 = np.linspace(0,15,100)
plt.figure(figsize = (16,10))
plt.plot(x_1, p_1(x_1))
plt.plot(x_1, f(x_1))
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Аппроксимация полиномом первой степени', fontsize = '24')

```

## Приближение многочленом второй степени


```python
x_2 = np.array([1, 8, 15])
A_2 = np.array([[1,1,1],[1,8,8**2], [1,15,15**2]])
b_2 = np.zeros(3)
b_2 = f(x_2)
w_2 = np.zeros(3) #polynomial coefficients 
w_2 = np.linalg.solve(A_2, b_2)
reversed_w_2 = w_2[::-1]
p_2 = np.poly1d(reversed_w_2) #polynom
print(p_2)
```

```python
x_2 = np.linspace(0,15,100)
plt.figure(figsize = (16,10))
plt.plot(x_2, p_2(x_2))
plt.plot(x_2, f(x_2))
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Аппроксимация полиномом второй степени', fontsize = '24')
```

## Приближение многочленом третьей степени

```python
x_3 = np.array([1, 8,10, 15])
A_3 = np.array([[1,1,1,1],[1,8,8**2, 8**3], [1, 10, 100, 1000], [1,15,15**2, 15**3]])
b_3 = np.zeros(4)
b_3 = f(x_3)
w_3 = np.zeros(4) #polynomial coefficients 
w_3 = np.linalg.solve(A_3, b_3)
reversed_w_3 = w_3[::-1]
p_3 = np.poly1d(reversed_w_3) #polynom
print(p_3)
```
